```python
#STRIP -REMOVE LEADING AND TRAILING WHITESPACES
s="   Hello   World"
print(s.strip())
print(s.lstrip())
print(s.rstrip())
s="###python###"
print(s.strip('#'))
```

```python
#SPLIT-USE TO SPLIT A STRING INTO A LIST
s="python is programming Language"
print(s.split())
print(s.split(' ',2))
print(s.rsplit(' ',2))
```

```python
#JOIN - USE TO JOIN THE ELEMENTS OF AN ITRERABLE
l=['h','e','l','l','o']
print(''.join(l))
l=['I','am','Nithish']
print(' '.join(l))
```

```python
#RELACE USED TO REPLACE A SPECIFIED STRING WITH ANOTHER STRING
s='I am virat and Iam Best batsman'
print(s.replace('virat','Dhoni'))
print(s.replace(' ','-',5))
```

```python
#UPPER-USED TO UPPER CHARACTER
#CAPITALIZE - USED TO CAPITALISE A FIRST LETTER IN A WORD
#TITLE - USE TO CAPITALISE THE FIRST WORD IN A SENTENCE
s="virat is a good batsman"
print(s.upper())
print(s.capitalize())
print(s.title())
s="DHONI IS A BEST CAPTAIN"
print(s.lower())
print(s.isupper())
print(s.islower())
```

```python
#isalpha,isnumeric,isalnum
s="Hello2"
print(s.isalpha())
a="12223"
a1="-12223"
a2="122/23"
a3="1222.3"
print(a.isnumeric())
print(a1.isnumeric())
print(a2.isnumeric())
print(a3.isnumeric())
b="NITHISH1701"
print(b.isalnum())
```

```python
#count()
c="python is a best programming language"
print(c.count('a'))
```
