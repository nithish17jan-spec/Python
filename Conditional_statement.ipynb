```python
#CONDITIONAL STATEMENT
age = 18

if age >= 18:
    print("You are eligible to vote")
```

```python
age = 16

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")
```

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```
