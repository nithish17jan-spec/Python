```python
# EB Bill Calculator

try:
    units = float(input("Enter total units consumed: "))

    if units < 0:
        print("Units cannot be negative.")
    else:
        bill = 0
        fixed_charge = 50.0

        # Slab calculation logic
        if units <= 100:
            bill = units * 1.5
        elif units <= 200:
            bill = (100 * 1.5) + ((units - 100) * 2.5)
        elif units <= 300:
            bill = (100 * 1.5) + (100 * 2.5) + ((units - 200) * 4.0)
        else:
            bill = (100 * 1.5) + (100 * 2.5) + (100 * 4.0) + ((units - 300) * 5.0)

        total_amount = bill + fixed_charge

        print("---------------------------")
        print(f"Units Consumed: {units}")
        print(f"Energy Charge: Rs. {bill:.2f}")
        print(f"Fixed Charge: Rs. {fixed_charge:.2f}")
        print(f"Total Bill: Rs. {total_amount:.2f}")
        print("---------------------------")

except ValueError:
    print("Invalid input. Please enter a numerical value.")
```
