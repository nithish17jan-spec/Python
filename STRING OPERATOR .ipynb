STRING OPERATOR IN PYHTON

```python
#Concatenate string
a="Hello"
b="ECE"
print(a+b)
```

```python
#REPEITION OPERATOR
a="NK"
print(a*7)
```

```python
#MEMBERSHIP OPERATOR
print("py"in "python")
print("yp"in "python")
```

```python
#Negative Mebership
print("x" not in "python")
print("py" not in "python")
```

```python
#Equality Operator
print("python"=="python")
print("10"==10)
```

```python
#Comparison Operator
print("Python">="python")
print("Python">"python")
print("Python"<"python")
print("Python"<="python")
print("Python"!="python")
```
