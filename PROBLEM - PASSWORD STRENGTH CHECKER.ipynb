```python
password = input("Enter your password:")

has_upper = False
has_lower = False
has_digit = False

if len(password) >= 8:
    for ch in password:
        if 'A' <= ch <= 'Z':
            has_upper = True
        elif 'a' <= ch <= 'z':
            has_lower = True
        elif '0' <= ch <= '9':
            has_digit = True

    if has_upper and has_lower and has_digit:
        print("Strong")
    elif (has_upper and has_lower) or (has_lower and has_digit):
        print("Medium")
    else:
        print("Weak")
else:
    print("Weak")
```
