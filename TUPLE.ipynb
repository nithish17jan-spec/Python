```python
#Tuple
t = (10, 20, 30)
```

```python
#create
t1 = ()                 # Empty tuple
t2 = (10,)              # Single element tuple (comma is must)
t3 = (1, 2, 3)
t4 = tuple([4, 5, 6])   # From list
```

```python
#access
t = (10, 20, 30, 40)

print(t[0])     # 10
print(t[-1])    # 40
```

```python
#slicing
print(t[1:3])    # (20, 30)
print(t[:2])     # (10, 20)
print(t[::-1])   # Reverse tuple
```
