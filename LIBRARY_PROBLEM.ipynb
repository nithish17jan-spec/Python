```python
#include <stdio.h>

struct Book {
    int id;
    char title[50];
    char author[50];
    int price;
};

// Function to read book details
void readBooks(struct Book b[], int n) {
    for (int i = 0; i < n; i++) {
        scanf("%d %s %s %d",
              &b[i].id,
              b[i].title,
              b[i].author,
              &b[i].price);
    }
}

// Function to display books with price > given amount
void displayBooks(struct Book b[], int n, int amount) {
    for (int i = 0; i < n; i++) {
        if (b[i].price > amount) {
            printf("%s\n", b[i].title);
        }
    }
}

int main() {
    int n, amount;
    struct Book b[50];

    scanf("%d", &n);        // number of books
    readBooks(b, n);        // read book details
    scanf("%d", &amount);   // price limit

    displayBooks(b, n, amount);  // display result

    return 0;
}
```
