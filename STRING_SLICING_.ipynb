```python
#SLICING
s="PYTHON PROGRAMMING LANGUAGE"
print(s[0:7])
print(s[:7])
print(s[0:])
print(s[0:8:2])
```

```python
#Negative third parameter
s="PYTHON PROGRAMMING LANGUAGE"
print(s[15:0:-1])
```

```python
#Reverse a string
s="PYTHON PROGRAMMING LANGUAGE"
print(s[::-1])
```
