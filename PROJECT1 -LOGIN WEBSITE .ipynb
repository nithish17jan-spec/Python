```python
# LOGIN WEBSITE
print("WELCOME TO THE APP")
print("1. LOG IN      2. SIGN UP")
choice = int(input("Enter your choice: "))

# Using a dictionary to store username:password pairs
user_db = {}

if choice == 2:
    u = input("Enter your username: ")
    while True:
        p = input("Create a password (must be alphanumeric, no spaces): ")

        # Check for spaces
        if " " in p:
            print("No spaces are allowed. Try again.")
            continue

        # Check if alphanumeric (letters and numbers only)
        if p.isalnum():
            user_db[u] = p
            print("Sign Up Successful!")
            break
        else:
            print("Password must only contain numbers and letters.")

elif choice == 1:
    u1 = input("Enter your username: ")
    if u1 in user_db:
        p1 = input("Enter your password: ")
        if user_db[u1] == p1:
            print("LOG IN SUCCESSFUL")
        else:
            print("Incorrect password.")
    else:
        print("Username not found.")
else:
    print("Invalid choice.")
```
